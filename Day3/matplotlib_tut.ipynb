# Introduction to Matplotlib

Matplotlib is a visualization library in Python for 2D plots of arrays.  
Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy(you'll get to know more about this is the further sessions) stack.   
One huge advantage of visualization is that it allows us visual access to large amounts of data in easily digestible visuals, thus making it easier to assume or establish relationships between various parameters involved in the plot.  
Matplotlib consists of several plots like line, bar, scatter, histogram etc.

## Essential Plots in matplotlib  

Matplotlib comes with a wide variety of plots. Plots helps to understand trends, patterns, and to make correlations. They’re typically instruments for reasoning about quantitative information. Some of the sample plots are covered here.

### Line Graphs


```python
# x-axis values 
x = [4, 2, 0, 6, 9] 
# Labelling X axis
plt.xlabel("X - Axis")

# Y-axis values 
y = [6, 9, 4, 2, 0] 
# Labelling Y axis
plt.ylabel("Y - Axis")

# Limits on X and Y axis
plt.axis([0, 10, 0, 10])

# Function to plot , ro arg for red circles
plt.plot(x,y,'ro')
  
# function to show the plot 
plt.show() 
```

### Histograms


```python
# Y-axis values 
y = [1, 3, 6, 7, 9] 
  
# Function to plot histogram 
plt.hist(y) 
  
# Function to show the plot 
plt.show()
```

### Scatter Plots

```python
# x-axis values 
x = [4, 2, 0, 6, 9] 
  
# Y-axis values 
y = [6, 9, 4, 2, 0] 
  
# Function to plot scatter 
plt.scatter(x, y) 
  
# function to show the plot 
plt.show() 
```

### Bar Graphs

```python
# x-axis values 
x = [6, 9, 4, 2, 0] 
  
# Y-axis values 
y = [4, 2, 0, 6, 9] 
  
# Function to plot the bar 
plt.bar(x,y) 
  
# function to show the plot 
plt.show() 
```

## Formatting graphs

For every x, y pair of arguments in the plot function, there is an optional third argument which is the format string that indicates the color and line type of the plot. The letters and symbols of the format string are from MATLAB, and you concatenate a color string with a line style string. The default format string is 'b-', which is a solid blue line.The code cell below contains the plot by using green triangles:

```python
plt.plot([2, 3, 4], [4, 9, 16], 'g^')
plt.axis([0, 7, 0, 19])
plt.show()
```

If matplotlib were limited to working with lists, it would be fairly useless for numeric processing as they are immutable. Generally, you will use numpy arrays. In fact, all sequences are converted to numpy arrays internally. The example below illustrates plotting several lines with different format styles in one command using arrays.

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red squares, blue dashes and green circles
plt.plot(t, t, 'rs', t, t**2, 'b--', t, t**3, 'go')
plt.show()
```

## Matplotlib object Hierarchy

A huge concept of matplotlib is the matplotlib object hierarchy.  
In the previous code cells, we have called plt.plot(x,y). This one-liner hides the fact that a plot is really a hierarchy of nested Python objects. A “hierarchy” here means that there is a tree-like structure of matplotlib objects underlying each plot.  
  
A Figure object is the outermost container for a matplotlib graphic, which can contain multiple Axes objects. One source of confusion is the name: an Axes actually translates into what we think of as an individual plot or graph (rather than the plural of “axis,” as we might expect).
  
You can think of the Figure object as a box-like container holding one or more Axes (actual plots). Below the Axes in the hierarchy are smaller objects such as tick marks, individual lines, legends, and text boxes. Almost every “element” of a chart is its own manipulable Python object, all the way down to the ticks (reference points on a scale) and labels:

Given below are two examples (with different methodologies) that details the usage of the subplots() function:

```python
# The two integer arguments to this function specify the number of rows and columns of the subplot grid
# The function returns a figure object and a tuple containing axes objects equal to nrows*ncols 
fig,a =  plt.subplots(2,2)
x = np.arange(1,5)
#Accessing each of the subplots through its index
a[0][0].plot(x,x*x*x)
a[0][0].set_title('cube')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('square root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exp')
a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')
plt.show()
```

```python
# making two correlated arrays that are drawn from discrete uniform 
x = np.random.randint(low=1, high=11, size=50)
y = x + np.random.randint(1, 5, size=x.size)
#column_stack makes x and y columns 1 and 2 respectively of the 2D array data
data = np.column_stack((x, y))

# figsize arg takes in the length and breadth of a subplot
# Here, ax is an 2D array of AxeSubplot with a shape (1,2)
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(8, 4))
# flattening the array
ax1,ax2 = ax.flatten()

# Another way to call scatter plots by using keywords 
ax1.scatter(x=x, y=y, marker='^', c='g', edgecolor='r')
# Dollar signs for italic formatting
ax1.set_title('Scatter: $x$ versus $y$')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')

# Another way to call the histogram plots via keywords 
# Bins are the blocks ina histogram that we use to combine values before we get the frequency
ax2.hist(data, bins=np.arange(data.min(), data.max()),label=('x', 'y'))
ax2.hist()
#legend function(loc denotes the location of the legend)
ax2.legend(loc=(0.65, 0.8))
ax2.set_title('Frequencies of $x$ and $y$')
#Shift the ticks to the right side
ax2.yaxis.tick_right()
```

Another way to call the subplot() function is given in the example below which also illustrates categorical plotting:   

```python
name = ['ones', 'tens', 'hundreds']
val = [1, 10, 100]

plt.figure(figsize=(8, 3))

plt.subplot(131) # 1 row 3 columns 1st figure
plt.bar(name, val)
plt.subplot(132) # 1 row 3 columns 2nd figure
plt.plot(name, val)
plt.subplot(133) # 1 row 3 columns 3rd figure
plt.scatter(name, val)
plt.suptitle('Categorical Plotting')
plt.show()
```

## Non-Linear axes

pyplot supports not only linear axis scales, but also logarithmic and logit ( log(x/1-x) ) scales. This is commonly used if data spans many orders of magnitude.   
This can be done by using the following line:  
- plt.xscale('log')  
- plt.yscale('logit')  
  
Other arguments that can be used are 'linear'(default), 'symlog'etc.

```python
x = np.random.randint(low=4, high=10, size=5000)
y = x + np.random.randint(3, 4, size=x.size)

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)


# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)


# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, 1/y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

# Adjusting the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.35,
                    wspace=0.35)
plt.show()
```
